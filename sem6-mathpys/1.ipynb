```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import math
```

```python
a = 3

def phi(x):
    return np.where(np.abs(x) <= 1 , np.cos(1.5 * x), 0)

def psi(x):
    return 0
```

```python
def u(x, t):
    return 0.5 * (phi(x - a*t) + phi(x + a*t))
```

```python
x = np.linspace(-2.1, 2.1, num=10000)

ts = [ 1/(4*a), 1/(2*a), 3/(4*a), 1/a, 2/(a) ]

for t in ts:
    plt.plot(x, u(x,t))
    
plt.legend(np.round(ts, 2))
plt.title("Профиль бесконечной струны")
plt.savefig('wave.png', dpi=200)
plt.show()
```
