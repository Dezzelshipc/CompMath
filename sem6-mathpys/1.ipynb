```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import math
```

```python
a = 3

def phi(x):
    return np.where(x <= 1 , np.where(-1 <= x , np.cos(1.5 * x), 0) , 0) 

def psi(x):
    return 0
```

```python
def u(x, t):
    return 0.5 * (phi(x - a*t) + phi(x + a*t))
```

```python
x = np.linspace(-2, 2, num=10000)
t = 1.5/a


fig, ax = plt.subplots()

plot, = plt.plot(x, u(x, 0))

def init():
    return plot,

def update(frame):
    plot.set_data(x, u(x, frame))
    return plot,

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2/a, 128),
                    init_func=init, blit=True)

plt.show()
```
