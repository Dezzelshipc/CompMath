```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
```

```python
l = 2
```

```python
r = expon.rvs(size=[100, 10], scale = 1/l)

r_mean = np.array([f3(ri, 10) for ri in r])

print(sum(r_mean)/len(r_mean))
```

```python
def f1(x, n):
    return np.mean(x)

def f2(x, n):
    return np.log(2) / np.median(x)

def f3(x, n):
    return 1 / np.mean(x)

def f4(x, n):
    return 1/(np.log(n)/n + np.mean(x))
```

```python
def plot(func, range, marker='co'):
    for i in range:
        in_one = i
        count = 100

        r = expon.rvs(size=[count, in_one], scale = 1/l)

        r_mean = np.array([func(ri, in_one) for ri in r])
        for v in r_mean:
            plt.plot(in_one, v, marker, markersize=1)
```

```python
plt.plot(-100,l, 'ko')
plt.plot(-100,l, 'ro')
plt.plot(-100,l, 'bo')
plt.plot(-100,l, 'yo')
plt.legend(["1: mean", "2: log2/median", "3: 1/mean", "4: 1/(logn / n + mean)"])

a, b = 20, 1000
rr = range(a, b, 10)

plot(f1, rr, 'ko') # black
plot(f2, rr, 'ro') # red
plot(f3, rr, 'bo') # blue
plot(f4, rr, 'yo') # yellow

plt.plot([0,b],[l,l], 'g-')
plt.plot([0,b],[1/l,1/l], 'c-')
plt.xlim([-10, b+10])

plt.show()
```
